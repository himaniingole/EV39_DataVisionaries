```r
install.packages(c("tidyverse", "ggplot2", "scales", "lubridate"))

library(tidyverse)
library(ggplot2)
library(scales)
library(lubridate)
```

```r
raw <- read.csv("/content/recalls_recall_listing.csv",
                stringsAsFactors = FALSE)

# Basic checks
head(raw)
str(raw)
dim(raw)
colnames(raw)
```

```r
missing_before <- colSums(is.na(raw))

missing_df <- data.frame(
  column = names(missing_before),
  missing_count = missing_before
)

ggplot(missing_df,
       aes(x = reorder(column, -missing_count),
           y = missing_count)) +
  geom_bar(stat = "identity", fill = "red") +
  coord_flip() +
  labs(
    title = "Missing Values per Column (Before Cleaning)",
    x = "Column",
    y = "Missing Count"
  )
```

```r
dir.create("visualizations", showWarnings = FALSE)
dir.create("visualizations/preprocessing_plots",
           recursive = TRUE,
           showWarnings = FALSE)

ggsave("visualizations/preprocessing_plots/missing_values_before.png",
       width = 8, height = 6)
```

```r
data <- raw %>%
  rename(
    recall_id      = Recall.Number,
    recall_date    = Date,
    product_name   = Name.of.product,
    hazard         = Hazard.Description,
    remedy         = Remedy,
    units_sold     = Units,
    incidents      = Incidents,
    manufacturer   = Manufacturers,
    importer       = Importers,
    distributor    = Distributors,
    country_origin = Manufactured.In
  )
```

```r
data_clean <- data %>%
  select(
    recall_id,
    recall_date,
    product_name,
    hazard,
    remedy,
    units_sold,
    incidents,
    manufacturer,
    importer,
    distributor,
    country_origin
  )
```

```r
data_clean$recall_date <- parse_date_time(
  data_clean$recall_date,
  orders = c("ymd", "mdy", "dmy", "BdY")
)

data_clean$recall_year <- year(data_clean$recall_date)
```

```r
data_clean$manufacturer[is.na(data_clean$manufacturer)] <- "Unknown"
data_clean$importer[is.na(data_clean$importer)] <- "Unknown"
data_clean$distributor[is.na(data_clean$distributor)] <- "Unknown"
data_clean$country_origin[is.na(data_clean$country_origin)] <- "Unknown"

data_clean$units_sold <- as.numeric(data_clean$units_sold)
data_clean$incidents  <- as.numeric(data_clean$incidents)

data_clean$units_sold[is.na(data_clean$units_sold)] <- 0
data_clean$incidents[is.na(data_clean$incidents)] <- 0
```

```r
data_clean <- data_clean[!duplicated(data_clean$recall_id), ]
```

```r
# Severity score
data_clean$severity_score <- data_clean$incidents * 2

# Hazard severity label (THIS WAS MISSING BEFORE)
data_clean$hazard_severity <- ifelse(
  data_clean$severity_score >= 5,
  "High",
  "Low"
)
```

```r
missing_after <- colSums(is.na(data_clean))

missing_after_df <- data.frame(
  column = names(missing_after),
  missing_count = missing_after
)

ggplot(missing_after_df,
       aes(x = reorder(column, -missing_count),
           y = missing_count)) +
  geom_bar(stat = "identity", fill = "green") +
  coord_flip() +
  labs(
    title = "Missing Values After Cleaning",
    x = "Column",
    y = "Missing Count"
  )
```

```r
write.csv(data_clean,
          "cleaned_recall_data.csv",
          row.names = FALSE)
```
